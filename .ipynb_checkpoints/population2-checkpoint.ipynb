### 우리동네지역의 남녀 성별 인구 분포

```python
import csv

with open("./datas/gender.csv") as f:
    data = csv.reader(f)
    header = next(data)
    
    for row in data:
        #print(row)
        if '경기도' in row[0]:
            print(row[2],row[105]) # 남성 총 인구, 여성 총 인구
            break
    
```

```python
import csv
import matplotlib.pyplot as plt

with open("./datas/gender.csv") as f:
    data = csv.reader(f)
    header = next(data)
    
    m = [] # 연령별 남성인구
    fm = [] #연령별 여성인구
    
    for row in data:
        if '대치2동' in row[0]:
            for i in row[3:104]: # 0~100[101개]
                m.append(int(i)) # 숫자로 변환
            for i in row[106:]:
                fm.append(int(i))
                
#print(m)
#print(fm)

#그래프로 출력
plt.plot(m, 'red')
plt.plot(fm, 'blue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

with open("./datas/gender.csv") as f:
    data = csv.reader(f)
    header = next(data)
    
    m = [] # 연령별 남성인구
    fm = [] #연령별 여성인구
    name = input("찾고 싶은 지역명을 입력하세요~")
    
    for row in data:
        if name  in row[0]:
            for i in row[3:104]: # 0~100[101개]
                m.append(-int(i.replace(",",""))) # 숫자로 변환
            for i in row[106:]:
                fm.append(int(i.replace(",","")))
            break
                
#print(m)
#print(fm)

#그래프로 출력
plt.rc("font", family="MBatang")
plt.title(name + "지역의 남여 성별 인구 분포")
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m, label='남성') #가로 막대형 그래프
plt.barh(range(101), fm, label='여성')
plt.legend()
plt.show()
```
